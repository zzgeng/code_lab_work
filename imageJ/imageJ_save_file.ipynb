## This code is designed to generate a code for imageJ macro, which can automatically split the channels and save each chennel as individual imageJ. 

The input picture format is .dv. The folder directly copied from microscope only contains .dv image file and .log files. 

The output of second section can be paste into imageJ macro and run!

```python
from os import walk 
import pandas as pd

dir_input = "/Users/zhuangzhuanggeng/Desktop/HBO/slice_image/01232022/"
filenames = pd.Series(next(walk(dir_input), (None, None, []))[2])

new_list = filenames[filenames.str.contains('.dv$')]
new_list
```

```python
dir_output = "/Users/zhuangzhuanggeng/Desktop/HBO/slice_image/01232022/imageJ/"

for x in new_list:
    print('open("' + str(dir_input) + str(x)+ '");')
    print('selectWindow("' + str(x) + ' - C=0");')
    print('resetMinAndMax();')
    print('saveAs("Jpeg", "' + str(dir_output) + str(x)+'_0.jpg");')
    print('close();')
    print('selectWindow("' + str(dir_output) + str(x) + ' - C=1");')
    print('resetMinAndMax();')
    print('saveAs("Jpeg", "' + str(x)+'_1.jpg");')
    print('close();')
    print('selectWindow("' + str(x) + ' - C=2");')
    print('resetMinAndMax();')
    print('saveAs("Jpeg", "' + str(dir_output) + str(x)+'_2.jpg");')
    print('close();')
```
